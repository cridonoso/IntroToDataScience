## Building our first figure 

We always use the `pyplot` module in `matplotlib`. It provides a MATLAB-like way of plotting: openning figures on your screen and managing figures GUIs

```python
import matplotlib.pyplot as plt
# util for working in jupyter notebook
%matplotlib inline
```

There are many ways to create charts. The most versatile way is to use `plt.subplots`. 

Using `plt.subplots` we can create from 1 to N `axes`

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
```

Now it is time to add labels and charts into the figure.

```python
axes[0].set_title('First ax')
axes[1].set_title('Second ax')
```

Notice python did not render the figure. To display the figure we can simply type `fig` which is the name of our figure object

```python
fig
```

```python
axes[0].set_xlabel('x label')
axes[0].set_ylabel('y label')

text_attr = fig.text(0.5, -0.05, 'common X', ha='center')
fig
```

```python
import numpy as np
```

```python
x = np.random.normal(0, 1, [10, 10])
```

```python
axes[1].plot(x[:, 0], x[:, 1])
fig
```

The most common plots are:
- **plot**: simple line x vs y
- **scatter**: independent points
- **hist**: histogram
- **bar**
- **imshow**: image plotting

```python
axes[0].imshow(x)
fig
```

Use `legend()` to display labels in data

```python
axes[1].plot(x[:, 0], x[:, 2], label='2nd dimension')
axes[1].legend()
fig
```

Finally, we can save figures in different formats using [`savefig()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html#matplotlib-pyplot-savefig) function 

```python
fig.savefig('images/my_first_figure.pdf', format='pdf')
fig.savefig('images/my_first_figure.png', format='png')
```

add `bbox_inches='tight'` into `savefig()` if something is cutted

```python
fig.savefig('images/my_first_figure.png', format='png', bbox_inches='tight')
```
